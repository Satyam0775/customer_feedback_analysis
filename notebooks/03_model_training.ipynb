```python
# === Notebook 03: Model Training ===
import joblib
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
import os

# === Paths ===
model_dir = r"C:\Users\satya\Downloads\customer_feedback_analysis\task\models"
split_data_path = os.path.join(model_dir, "split_data.pkl")
model_save_path = os.path.join(model_dir, "feedback_model.pkl")

# === Load train-test splits ===
X_train_vec, X_test_vec, y_train, y_test = joblib.load(split_data_path)
print("âœ… Data splits loaded successfully.")

# === Train Logistic Regression model ===
model = LogisticRegression(max_iter=200)
model.fit(X_train_vec, y_train)
print("âœ… Model training completed.")

# === Save trained model ===
joblib.dump(model, model_save_path)
print(f"âœ… Model saved at: {model_save_path}")

# === Quick evaluation ===
preds = model.predict(X_test_vec)
print("\nðŸ“Š Quick Evaluation Report:\n")
print(classification_report(y_test, preds))
```
